```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
name_str='test_accuracy_{}.npz'
name_str_seed = 'test_accuracy_{}_seed_{}.npz'
sample_list = np.arange(100,600,100)

for sample in sample_list:
    data = np.load(name_str.format(sample))
    for i in range(7):
        data_seed = np.load(name_str_seed.format(sample, i))
        N_all_seeds_w1= np.append(N_all_seeds_w1, data_seed['w1'])
        N_all_seeds_w2 = np.append(N_all_seeds_w2, data_seed['w2'])
        #N_weights_w1 = np.append(N_weights_w1, data_seed['accuracy'])
        #N_weights_w2 = np.append(N_weights_w2, data_seed['accuracy'])
        print(data_seed['w2'])
    print('//')
    N = np.append(N, data['N'])
    #N_w1 = np.append(N_w1, data['w1'])
    #N_w2 = np.append(N_w2, data['w2'])
    #N_std_dev_w1 = np.append(N_std_dev_w1, data['error_bar_w1'])
    #N_std_dev_w2 = np.append(N_std_dev_w2, data['error_bar_w2'])
    #N_accuracy = np.append(N_accuracy, data['accuracy'])
```

```python

N_seed_w1 = N_all_seeds_w1.reshape(5,7)
#N_weights_w1 = N_weights_w1.reshape(5,7)
N_seed_w2 = N_all_seeds_w2.reshape(5,7)
#N_weights_w2=  N_weights_w2.reshape(5,7)
```

```python
N_mean_w1 = np.mean(N_seed_w1, axis=1)
N_std_w1 = np.std(N_seed_w2, axis=1)
N_mean_w2 = np.mean(N_seed_w2, axis=1)
N_std_w2 = np.std(N_seed_w2, axis=1)
```

```python
fig, ax = plt.subplots()
import seaborn as sns

sns.set(style='ticks', context='talk')
ax.scatter(N, N_mean_w1, color="k")
for (i,element) in enumerate(N_seed_w1):
    count1 = 0
    count2 = 0
    count3 = 0
    count4 = 0
    count5 = 0
    count6 = 0
    count7 = 0
    count8 = 0
    count9 = 0
    count10=0
    for number in element:
        if number == 0.07982583454281568:
            ax.scatter(x=N[i]+count1,y= number, color='b')
            count1 += 10
        elif number == 0.16690856313497823:
            ax.scatter(x=N[i]+count2,y= number, color='b')
            count2 += 10
        elif number == 0.12336719883889695:
            ax.scatter(x=N[i]+count3,y= number, color='b')
            count3 += 10
        elif number == 0.0362844702467344:
            ax.scatter(x=N[i]+count4,y= number, color='b')
            count4 += 10
        elif number == 0.26153846153846155:
            ax.scatter(x=N[i]+count5,y= number, color='b')
            count5 += 10
        elif number == 0.43570391872278663:
            ax.scatter(x=N[i]+count6,y= number, color='b')
            count6 += 10
        elif number ==    0.2641509433962264:
            ax.scatter(x=N[i]+count7,y= number, color='b')
            count7 += 10
        elif number == 0.43831640058055155:
            ax.scatter(x=N[i]+count8,y= number, color='b')
            count8 += 10
        else:
            ax.scatter(x=N[i], y=number, color='b')

ax.plot(N, 6*0.3445*np.sqrt(6)/(np.sqrt(784+N)), color = "k")
ax.errorbar(N, N_mean_w1, N_std_w1, capsize = 10.0, color= "k")
ax.set_xlabel("Number of Hidden Neurons (H)")
ax.set_ylabel("sigma_1")
ax.set_ylim(0,0.8)
ax.set_title("Sigma_1 vs. Number of Hidden Neurons (H)")
fig.savefig("test_W1_vs_N.png", bbox_inches = 'tight')
```
