```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# N = Number of Input Neurons
N = np.array([])
N_w1 = np.array([])
N_w2 = np.array([])
N_std_dev_w1 = np.array([])
N_std_dev_w2 = np.array([])
N_accuracy = np.array([])
```

```python
name_str='train_accuracy_{}.npz'
sample_list = np.arange(100,600,100)

for sample in sample_list:
    data = np.load(name_str.format(sample))
    N = np.append(data['N'], N)
    N_w1 = np.append(data['w1'], N_w1)
    N_w2 = np.append(data['w2'], N_w2)
    N_std_dev_w1 = np.append(data['error_bar_w1'], N_std_dev_w1)
    N_std_dev_w2 = np.append(data['error_bar_w2'], N_std_dev_w2)
    #N_accuracy = np.append(data['accuracy'], N_accuracy)
```

```python
fig, ax = plt.subplots()
ax.scatter(N, N_w1, color="k")
ax.errorbar(N, N_w1, N_std_dev_w1/0.3445, capsize = 10.0, color= "k")
ax.set_xlabel("Number of Hidden Neurons (N)")
ax.set_ylabel("Standard Deviation of W1")
fig.savefig("train_W1_vs_N.png")
```

```python
fig2, ax2 = plt.subplots()
ax2.scatter(N, N_w2, color="k")
ax2.errorbar(N, N_w2, N_std_dev_w2/0.3445, capsize = 10.0, color= "k")
ax2.set_xlabel("Number of Hidden Neurons (N)")
ax2.set_ylabel("Standard Deviation of W2")
fig2.savefig("train_W2_vs_N.png")
```
