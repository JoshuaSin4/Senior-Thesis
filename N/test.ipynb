```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# N = Number of Input Neurons
N = np.array([])
N_w1 = np.array([])
N_w2 = np.array([])
N_std_dev_w1 = np.array([])
N_std_dev_w2 = np.array([])
N_accuracy = np.array([])
N_all_seeds_w1 = np.array([])
N_all_seeds_w2 = np.array([])
N_weights_w1 = np.array([])
N_weights_w2 = np.array([])
```

```python
name_str='accuracy_{}.npz'
name_str_seed = 'accuracy_{}_seed_{}.npz'
sample_list = np.arange(100,500,50)

for sample in sample_list:
    data = np.load(name_str.format(sample))
    for i in range(10):
        data_seed = np.load(name_str_seed.format(sample, i))
        N_all_seeds_w1= np.append(N_all_seeds_w1, data_seed['w1'])
        N_all_seeds_w2 = np.append(N_all_seeds_w2, data_seed['w2'])
        #N_weights_w1 = np.append(N_weights_w1, data_seed['accuracy'])
        #N_weights_w2 = np.append(N_weights_w2, data_seed['accuracy'])
        print(data_seed['w2'])
    print('//')
    N = np.append(N, data['N'])
    #N_w1 = np.append(N_w1, data['w1'])
    #N_w2 = np.append(N_w2, data['w2'])
    #N_std_dev_w1 = np.append(N_std_dev_w1, data['error_bar_w1'])
    #N_std_dev_w2 = np.append(N_std_dev_w2, data['error_bar_w2'])
    #N_accuracy = np.append(N_accuracy, data['accuracy'])
```

```python

N_seed_w1 = N_all_seeds_w1.reshape(8,10)
#N_weights_w1 = N_weights_w1.reshape(5,7)
N_seed_w2 = N_all_seeds_w2.reshape(8,10)
#N_weights_w2=  N_weights_w2.reshape(5,7)
```

```python
N_mean_w1 = np.mean(N_seed_w1, axis=1)
N_std_w1 = np.std(N_seed_w2, axis=1)
N_mean_w2 = np.mean(N_seed_w2, axis=1)
N_std_w2 = np.std(N_seed_w2, axis=1)
```

```python
fig, ax = plt.subplots()
import seaborn as sns

sns.set(style='ticks', context='talk')
ax.scatter(N, N_mean_w1, color="k")
for (i,element) in enumerate(N_seed_w1):
    count1 = 0
    count2 = 0
    for number in element:
        if number == 0.725689404934688:
            ax.scatter(x=N[i]+count1,y= number, color='b')
            count1 += 5
        elif number == 0.43541364296081286:
            ax.scatter(x=N[i]+count2,y= number, color='b')
            count2 += 5
       
        else:
            ax.scatter(x=N[i], y=number, color='b')

ax.plot(N, np.sqrt(6)/(np.sqrt(N)+np.sqrt(N)), color = "k")
ax.errorbar(N, N_mean_w1, N_std_w1, capsize = 10.0, color= "k")
ax.set_xlabel("Number of Input Neurons (N)")
ax.set_ylabel("std_w1")

ax.set_title("Sigma_1 vs. Number of Input Neurons (N)")
fig.savefig("test_W1_vs_N.png", bbox_inches = 'tight')
```
